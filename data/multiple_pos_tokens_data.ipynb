```python
import random
import numpy as np

total_num_clauses = 96
ALPHABET = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
CLAUSES = list(map(str,list(range(total_num_clauses))))
num_vars = 10
num_clauses = 40
k = 3

def get_sample():
    sampled_letters = random.sample(ALPHABET, num_vars)

    sampled_clauses = random.sample(CLAUSES, num_clauses)
    final_str = sampled_letters + sampled_clauses
    letter_clauses = []
    for letter in final_str:
        letter_clauses.append(random.sample(sampled_clauses, k))
    # Convert list of lists to numpy array and pad with zeros if needed
    # Create list of lists for clauses and their letters
    clause_letters = []
    for clause in sampled_clauses:
        letters = []
        for i, letter_clause_list in enumerate(letter_clauses):
            if i > num_vars-1:
                break
            if clause in letter_clause_list:
                letters.append(final_str[i])
        clause_letters.append(letters)

    letter_clauses_matrix = np.array(letter_clauses).T
    # Create k x num_vars matrix of ones
    ones_matrix = np.ones((k, num_vars))
    zeros_matrix = np.zeros((k, num_clauses))
    labels1 = np.zeros((len(ALPHABET), num_vars))
    labels2 = []
    for i,clause in enumerate(sampled_clauses):
        empty = np.zeros(len(ALPHABET))
        contained = clause_letters[i]
        for i,a in enumerate(ALPHABET):
            if a in contained:
                empty[i] = 1
        labels2.append(empty)
    labels2 = np.array(labels2).T
    # Concatenate labels1 and labels2 horizontally
    labels = np.concatenate((labels1, labels2), axis=1)

    # Concatenate ones_matrix and zeros_matrix horizontally
    mask_matrix = np.concatenate((ones_matrix, zeros_matrix), axis=1)

    return {
        'labels':labels.astype(np.int64),
        'masks':mask_matrix.astype(np.int64),
        'tokens':final_str, 
        'pos_embeddings': letter_clauses_matrix.astype(np.int64)

    }


valid = {}
train = {}
num_val = 1000
num_train = 20000
for i in range(num_val):
    valid[i] = get_sample()

for i in range(num_train):
    train[i] = get_sample()


import pickle
with open("test.pkl", "wb") as f:
    pickle.dump(valid, f)

with open("train.pkl", "wb") as f:
    pickle.dump(train, f)
```
